# 04. Create a bar chart of the top 20 starting stations

```python
df.head()
```

```python
#Add a new column named 'value' and set every entry in this column to 1
df['value'] = 1
```

```python
df.head()
```

```python
#Groupby **start_station_name** and count the number of entries in each group. **the count are the number of occurrences of each **start_station_name**
df_grouped = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
#Check
df_grouped.head()
```

```python
#Aggregate the occurrences of the stationsâ€™ names to select the top 20 most visited stations 
top20 = df_grouped.nlargest(20, 'value')
```

```python
#Check
top20
```

```python
#Create a barchart of the top 20 starting stations
palette = sns.light_palette("#4682B4", n_colors=20, reverse=True) #Set the color choice

sns.barplot(data = top20, x = 'value', y = 'start_station_name', palette=palette) #OR palette=('Blues_r')
plt.title("Most Frequent Start Stations - Top 20")

#Save the plot
plt.savefig('Top20_Start_Stations.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```

# Observations

+ The color palette I initially chose doesn't work for this plot.
+ The bar chart needs 20 distinct colors, one for each station or column.
+ The default setting of the palette does not contain enough colors.
+ I choose a light blue palette, where the darkest shade represents the most frequently visited station, and the lightest shade corresponds to the least visited station in the top 20 list.

```python
df.shape
```

```python
#Convert the data column to type date -just the date component
df['started_at'] = pd.to_datetime(df['started_at'], format = '%Y-%m-%d')

#df['started_at'] = pd.to_datetime(df['started_at']).dt.date
```

```python
df.dtypes
```

```python
df.head(3)
```

```python
#Sort the dataframe by the date
df.sort_values(by='started_at', inplace=True)
```

```python
df.head(3)
```

```python
#Set the date column to be the index of the dataframe **prepare for plotting**
df_dualaxis = df.set_index('started_at')
```

```python
df_dualaxis.head(3)
```

```python
#Create the dual-axis line plot using Seaborn **The function **sns.lineplot()** differs from Matplotlib**

#Create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(10, 5))

#Make a plot of the bike rides
sns.lineplot(data = df_dualaxis['daily_bike_rides'], color = "navy",ax=ax)
ax.set_xlabel("Year 2022", fontsize = 12) # set x-axis label
ax.set_ylabel("Daily Bike rides",color = "navy", fontsize = 14) # set y-axis label

#Create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

#Make a plot with the temperatues on the y-axis using second axis object
sns.lineplot(data = df_dualaxis['avgTemp'], color="r", ax=ax2)
ax2.set_ylabel("Average temperatures", color = "red",fontsize=12)
plt.title('Temperature and trips in 2022', fontsize = 14)

#Save the plot
plt.savefig('Seaborn_dual_axis.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```

# 06. A Boxplot of one of the categorical variables in your data set

```python
df.dtypes
```

```python
df.head()
```

```python
#Create a box plot of the **rideable_type** a categorical variable
fig,ax = plt.subplots(figsize=(10, 5))

sns.boxplot(x = 'rideable_type', y = 'avgTemp', data = df, palette = {'classic_bike': '#2E8B57', 'electric_bike': '#4682B4', 'docked_bike': '#A9A9A9'})
ax.set_xlabel('Bike Type', fontsize=12) # set x-axis label
ax.set_ylabel('Average Temperature', fontsize=12) # set x-axis label
plt.title('Bike Type and Temperature in 2022', fontsize = 14)

#Save the plot
fig.savefig('Boxplot_BikeType_Temperature.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()

# Set the x-axis label using Matplotlib
#plt.xlabel("Bike Type", fontsize=12) #can use fontweight='bold'
```

# Observations

+ The range of data points is similar across all three bike types.
+  The medians for classic and electric bikes are very similar, while the median for docked bikes is slightly higher.
+ The three box plots reveal outliers at the lower end, with docked bikes having the most.
______

+ I chose to analyze the relationship between rideable_type and avgTemp to see if temperature affects the type of bike used.
+ There are three bike types: classic, electric, and docked bikes.
+ The box plot shows little difference between classic and electric bikes.
+ However, docked bikes differ significantly from the other types.
+ Docked bikes have many outliers when temperatures drop.
+ This suggests that bike rentals decrease in NYC when it's cold.
+ Docked bikes show a higher median, indicating most rentals occur when temperatures are below 22Â°C.

# Create a FacetGrid

```python
#Create a grid object on the df, split where the rideable_type
grid = sns.FacetGrid(df, col = "rideable_type")

#Use the grid.map function to have a histogram of the column avgTemp 
grid.map(sns.histplot, 'avgTemp', bins = 15)
plt.show()
```

```python
#Define custom palette - based on the ride type
palette = {
    'classic_bike': '#2E8B57',  #Green
    'electric_bike': '#4682B4', #Blue
    'docked_bike': '#A9A9A9'    #Gray
}
#Create a FacetGrid
grid = sns.FacetGrid(df, col="rideable_type", hue="rideable_type", palette=palette)

#Map plots to the FacetGrid
grid.map(sns.histplot, 'avgTemp', bins = 15)

#Save the plot
plt.savefig('FacetGrid.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```

# Observations

+ Inthe graph above, the FacetGrid plot allows us to visualize data distribution based on average temperature.
+ We observe that classic bikes are more popular than electric bikes, and rentals increase with rising temperatures.
+ The count of classic bike rentals varies greatly, ranging from 20,000 to 10,000, whereas electric bike rentals show less variation, remaining below 20,000.
+ The distribution pattern is similar across the two ride types.
------
+ This finding aids the analysis by revealing insights into bike demand. Different types are more popular than others.
+ Stakeholders should consider reallocating rideable types based on station demand, adding more classic bikes where electric bikes are less popular, and placing electric bikes in stations where they are in higher demand.
